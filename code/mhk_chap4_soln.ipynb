```python
import random
import thinkstats2
import thinkplot

%matplotlib inline
```

```python
randList = [random.random() for x in range(1000)]
```

```python
pmf = thinkstats2.Pmf(randList, label='rand')
thinkplot.Pmfs([pmf])
thinkplot.Show(xlabel='random number',
               ylabel='PMF',
               axis=[-0.1,1.1,0.0,0.0011],
               title='PMF for random numbers')
```

```python
cdf = thinkstats2.Cdf(randList, label='rand')
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='random number', ylabel='CDF', title='CDF for random numbers')
```

randList is a list of 1000 random numbers where each random number has an equal probabilty of being part of the 1000 numbers generated. This the reason why PMF plot is a uniform distribution which means that CDF plot will be a straight line (as shown above).
